```python
def symbolToNumber(symbol):
    if symbol == "A":
        return 0
    elif symbol =="C":
        return 1
    elif symbol == "G":
        return 2
    elif symbol == "T":
        return 3
```

```python
def patternToNumber(pattern):
    lenP = len(pattern)
    if lenP == 0:
        return 0
    symbol = pattern[lenP-1]
    prefix = pattern[0:lenP-1]
    return 4*patternToNumber(prefix) + symbolToNumber(symbol)
    
    
        
```

```python
def numberToSymbol(index):
    if index == 0:
        return "A"
    elif index == 1:
        return "C"
    elif index == 2:
        return "G"
    elif index == 3:
        return "T"
```

```python
def numberToPattern(number,k):
    if k == 1:
        return numberToSymbol(number)
    prefixIndex = int(number/4)
    lastIndex = number % 4
    return numberToPattern(prefixIndex, k-1) + numberToSymbol(lastIndex)
    
    
```

```python
def neighbors(pattern,d):
    if d == 0:
        return pattern
    if len(pattern) == 1:
        return {"A","C","G", "T"}
    neighborhood = set()
    suffixNeighbors = neighbors(pattern[1:len(pattern)], d)
    for suffixNeighbor in suffixNeighbors:
        if hammingDistance(suffixNeighbor, pattern[1:len(pattern)]) < d:
            for nucleotide in {"A", "C", "G", "T"}:
                neighborhood.add(nucleotide + suffixNeighbor)
        else:
            neighborhood.add(pattern[0]+suffixNeighbor)
    return neighborhood
    

            
                
    
```

```python
def frequentPatternsWithMismatch(genome,k,d):
    frequentPatterns = set()
    frequencyArray = []
    close = []
    for i in range(0, 4**k):
        close.append(0)
        frequencyArray.append(0)
    for i in range(0, len(genome)-k + 1):
        neighborhood = neighbors(genome[i:i+k],d)
        for pattern in neighborhood:
            index = patternToNumber(pattern)
            close[index] = 1
    for i in range(0, 4**k):
        if close[i] == 1:
            pattern = numberToPattern(i,k)
            frequencyArray[i] = approxPatternCount(pattern,genome,d)
    maxCount = max(frequencyArray)
    for i in range(0, 4**k):
        if frequencyArray[i] == maxCount:
            frequentPatterns.add(numberToPattern(i,k))
            print(numberToPattern(i,k), end=" ")
    return frequentPatterns
            
            
            
            
            
    
        
        
    
    
```

```python
def approxPatternCount(pattern, genome, d):
    count = 0
    for i in range(0, len(genome)-len(pattern)+1):
        apattern = genome[i: i + len(pattern)]
        if hammingDistance(apattern, pattern) <= d:
            count = count + 1
    return count
```

```python
def reverseComplement(text):
    reverse=""
    for b in text:
        if b == "A":
            reverse = "T" + reverse
        elif b == "T":
            reverse = "A" + reverse
        elif b == "C":
            reverse = "G" + reverse
        elif b == "G":
            reverse = "C" + reverse
    return reverse
```

```python
reverseComplement("ACG")
```

```python
def frequentPatternsWithMismatchesAndReverseComplement(genome,k,d):
    frequentPatterns=set()
    neighborhoods=list()
    for i in range(0, len(genome)-k+1):
        pattern = genome[i:i+k]
        rPattern= reverseComplement(pattern)
        neighborhoods.append(neighbors(pattern,d))
        neighborhoods.append(neighbors(rPattern,d))
    neighborhoodArray = []
    for neighborhood in neighborhoods:
        for neighbor in neighborhood:
            neighborhoodArray.append(neighbor)
    count = []
    index = []
    for i in range(0, len(neighborhoodArray)):
        pattern = neighborhoodArray[i]
        count.append(1)
        index.append(patternToNumber(pattern))
    sortedIndex = sorted(index)
    for i in range(0, len(neighborhoodArray)-1):
        if sortedIndex[i] == sortedIndex[i+1]:
            count[i+1] = count[i] + 1
    maxCount = max(count)
    for i in range(0, len(neighborhoodArray)):
        if count[i] == maxCount:
            pattern = numberToPattern(sortedIndex[i], k)
            frequentPatterns.add(pattern)
            print(pattern, end=" ")
    return frequentPatterns
```

```python
def skew(text):
    lenT = len(text)
    countC = 0
    countG = 0
    difference = [0]
    for i in range(0, lenT):
        if text[i] == "C":
            countC = countC + 1
        if text[i] == "G":
            countG = countG + 1
        difference.append(countG-countC)
    return difference

def minimizeSkew(genome):
    difference = skew(genome)
    minimum = min(difference)
    results =[]
    for i in range(0, len(difference)):
        if difference[i] == minimum:
            print (i)
            results.append(i)
```
